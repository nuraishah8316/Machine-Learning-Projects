```python
# Supervised Machine Learning
# Regression - Polynomial Regression
# Polynomial regression-increase the degree of independent variable to allow model to capture nonlinear relationship between X and Y

# Second Project : Predicting Salary Based On Level

# import all necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.linear_model as lm
lin_reg = lm.LinearRegression()
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Position_Salaries.csv')
data
```

```python
# Select Independent And Dependent Variable
X = data.Level.values.reshape(-1,1) # single independent variable, need to reshape to 2D
Y = data.Salary.values
```

```python
# Model Development
lin_reg.fit(X,Y)  # model use for future prediction
```

```python
data['predicted_salaries'] = lin_reg.predict(X)
```

```python
data[['Level','Salary','predicted_salaries']]
```

```python
pred_values = lin_reg.predict(X)
print(pred_values)
```

```python
# Plot Actual Values
plt.scatter(X,Y,color='red',label='Actual Values')
# Plot Predicted Values
plt.plot(X,pred_values,color='green',label='Predicted Values')
plt.title('Level Vs Salary')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
# Let's Try Using Normalization
import sklearn.preprocessing as pp
Y_nrmlz = pp.normalize([Y]).flatten() # use 'flatten() method to convert dependent variable from 2D to 1D
Y_nrmlz
```

```python
# model Development
lin_reg.fit(X,Y_nrmlz)
```

```python
# Plot Actual Values
plt.scatter(X,Y_nrmlz,color='red',label='Actual Values')
# Plot Predicted Values
plt.plot(X,lin_reg.predict(X),color='green',label='Predicted Values')
plt.title('Level Vs Salary With Normalization')
plt.xlabel('Level')
plt.ylabel('Salary With Normalization')
plt.legend()
plt.show()
```

```python
# Polynomial Regression
# Polynomial regression is type of regression analysis that models a non-linear
# relationship between the independent and dependent variables

# Generate Polynomial Features

# by default the degree of independent variable is 1
# but in Polynomial Regression we will increase the degree of Independent Variable
# and the we will Train Our Model

from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)

x_poly_2 = poly_reg.fit_transform(X)  # X is IV--> transform it --> degree = 1 to 2 (3 columns created)
x_poly_2

# when degree = 2, then new features will be created like this  1   x*1   x*x (3 column)
# when degree = 3, then new features will be created like this  1   x*1   x*x   x*x*x (4 column)
# when degree = 4, then new features will be created like this  1   x*1   x*x   x*x*x  x*x*x*x (5 column)
```

```python
# Train our model using Degree = 2
lin_reg.fit(x_poly_2,Y)
```

```python
# Plot Actual Values
plt.scatter(X,Y,color='red',label='Actual Values')
# Plot Predicted Values
plt.plot(X,lin_reg.predict(x_poly_2),color='green',label='Predicted Values')
plt.title('Level Vs Salary With Degree = 2')
plt.xlabel('Level')
plt.ylabel('Salary With Degree = 2')
plt.legend()
plt.show()
```

```python
# Train Our Model Degree = 3
poly_reg = PolynomialFeatures(degree=3)
x_poly_3 = poly_reg.fit_transform(X)  # X is IV--> transform it --> degree = 2 to 3 (4 columns created)
x_poly_3
# when degree = 3, then new features will be created like this  1   x*1   x*x   x*x*x (4 column)
```

```python
lin_reg.fit(x_poly_3,Y)
```

```python
# Plot Actual Values
plt.scatter(X,Y,color='red',label='Actual Values')
# Plot Predicted Values
plt.plot(X,lin_reg.predict(x_poly_3),color='green',label='Predicted Values')
plt.title('Level Vs Salary With Degree = 3')
plt.xlabel('Level')
plt.ylabel('Salary With Degree = 3')
plt.legend()
plt.show()
```

```python
# Train Our Model Degree = 4
poly_reg = PolynomialFeatures(degree=4)
x_poly_4 = poly_reg.fit_transform(X)  # X is IV--> transform it --> degree = 3 to 4
lin_reg.fit(x_poly_4,Y)
```

```python
# Plot Actual Values
plt.scatter(X,Y,color='red',label='Actual Values')
# Plot Predicted Values
plt.plot(X,lin_reg.predict(x_poly_4),color='green',label='Predicted Values')
plt.title('Level Vs Salary With Degree = 4')
plt.xlabel('Level')
plt.ylabel('Salary With Degree = 4')
plt.legend()
plt.show()
```

```python
# There's no uppper limit on the degree of a polynomial.
# even n - 1 where n is the number odf data points --> which would perfectly pass through all points
# The higher the degree, the more complex and wiggly your curve becomes.
# Rule of Thumb:
# Fewer than 100 points - try degree 2 or 3
# Medium data(hundreds=thousands) - try degree 2-5
# Large data - polymonial regression often replaced by other nonlinear models)like decision trees or neutral networks)
```

```python
# R-squared
# help you to know how well the explain the data
# or how well the model fits the data

from sklearn.metrics import r2_score
accuracy = r2_score(Y,lin_reg.predict(x_poly_4))
print('Regression Score Function :',accuracy)
print(f'Regression Score Function : {int(accuracy*100)}%')
```

```python
# Final Code For Polynomial Regression

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Position_Salaries.csv')
X = data[['Level']].values
y = data['Salary'].values

# Try polynomial degrees 1 to 5
plt.scatter(X, y, color='red', label='Actual Data')

for degree in range(1, 6):
    # Create polynomial features
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)

    # Fit linear regression
    lin_reg = LinearRegression()
    lin_reg.fit(X_poly, y)

    y_pred = lin_reg.predict(X_poly)

    # Calculate r2 score
    r2 = r2_score(y, y_pred)
    print(f"Degree {degree} - R2 Score: {r2}")

    # Plot the curve
    plt.plot(X, y_pred, label=f'Degree {degree}')

plt.title("Polynomial Regression")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
