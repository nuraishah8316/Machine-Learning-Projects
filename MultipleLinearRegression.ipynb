```python
# Third Project : Predicting Profit Based On Multiple Independent Variable

# Load The Dataset
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Startups_Data.csv')
data.head()
```

```python
data.info()
```

```python
# Select Independent and Dependent Variable
data.columns
```

```python
x = ['R&D Spend', 'Administration', 'Marketing Spend', 'State']   # create list (Independent Variable)
x = data[x].values  # to only generate values, not with list and column
x
```

```python
x.shape # Independent Variable already in 2D, does no need to reshape
```

```python
y = data.Profit.values  # dependent Variable
y
```

```python
# OneHotEncoding
# One-Hot Encoding is a data preprocessing technique used to convert
# categorical (text) data into numerical form so that machine learning models can
# understand it.


from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
transformer = ColumnTransformer(
    transformers = [('OneHot',OneHotEncoder(),[3])],  # List of (name,transformer,columns), List of tuples = [()]
    remainder = 'passthrough' # please do not remove non specified columns
)

x = transformer.fit_transform(x)
x = x.astype('int')
```

```python
x

# california  florida   newYork   r_d     admin     market
#     0,        0,          1,   165349,  136897,   471784
#     1,        0,          0,   162597,  151377,   443898

# newYork --> 0 0 1
# california --> 1 0 0
# florida --> 0 1 0
```

```python
# train test split
# in machine learning (ML), we teach computers to make predictions
# or decisions based on data.
# To make sure our model learns properly and doesn’t just memorize,
# we divide our data into two parts:
# 1. Training set – used to teach the model
# 2. Testing set – used to check if the model really learned or just memorized

# random_state do?
# random_state acts like a seed — it makes sure the random process happens
# the same way every time you run the code. That means if you (or someone else)
# run your code again, you’ll get exactly the same split.

# Why number 42?
# There’s nothing special about 42 mathematically — it’s just
# a popular “inside joke” among programmers, from a famous
# sci-fi book “The Hitchhiker’s Guide to the Galaxy”,
# where 42 is “the answer to life, the universe, and everything.”
```

```python
# train test split-make sure the ML is learning NOT memorizing
# 0.2 means 20% of the data will be using for Testing
# remaining 0.8 means 80% of the data will be using for Training
# maximum for Testing is 0.35, for Training to be higher

from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print('Original Shape')
print(x.shape)
print(y.shape)
print('After Splitting')
print(x_train.shape)    # 40 records use for training, 10 records will be used for training
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import sklearn.linear_model as lm
mlr = lm.LinearRegression()

mlr.fit(x_train,y_train)
```

```python
pred_values = mlr.predict(x_test) # testing the ml(x_test)
print(pred_values)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,pred_values)  # comparing the IV with dependent variable(y_test)
print(f'Regression Score : {score}')
```
